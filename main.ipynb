```python
import json
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer

# Load the FAQ data
with open(r"C:\Users\Moham\Projects\AI\Chat bot\aramex_faq.json", "r", encoding="utf-8") as f:
    faq_data = json.load(f)

# Initialize the embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Extract questions and answers
questions = [item['question'] for item in faq_data]
answers = [item['answer'] for item in faq_data]
metadata = faq_data

# Create embeddings for all questions
question_embeddings = model.encode(questions)
embedding_dim = question_embeddings.shape[1]

# Create FAISS index
index = faiss.IndexFlatL2(embedding_dim)
index.add(np.array(question_embeddings, dtype=np.float32))

# Save index
faiss.write_index(index, "aramex_faq_index.faiss")

print(f"✓ FAISS index created with {len(questions)} FAQ items")
print(f"✓ Embedding dimension: {embedding_dim}")

# Define retrieval function
def retrieve_answer(query_text, top_k=3):
    query_embedding = model.encode([query_text])
    distances, indices = index.search(query_embedding.astype(np.float32), top_k)
    
    # تحويل المسافات إلى قيم عكسية
    inverse_distances = 1 / (distances[0] + 1e-6)  # نضيف قيمة صغيرة لتجنب القسمة على صفر
    total = sum(inverse_distances)
    
    results = []
    for i, idx in enumerate(indices[0]):
        probability = inverse_distances[i] / total
        results.append({
            'question': metadata[idx]['question'],
            'answer': metadata[idx]['answer'],
            'distance': distances[0][i],
            'confidence': round(probability * 100, 2)  # كنسبة مئوية
        })
    return results

```

```python
# Test it
test_results = retrieve_answer("How do I track my shipment?", top_k=3)
for result in test_results:
    print(f"\nQ: {result['question']}")
    print(f"A: {result['answer']}")
    print(f"Confidence: {result['confidence']}%")
```

```python
import os
# Load resources
model = SentenceTransformer("all-MiniLM-L6-v2")   # same model you used for embeddings
index = faiss.read_index("aramex_faq_index.faiss")
with open(r"C:\Users\Moham\Projects\AI\Chat bot\aramex_faq.json", "r", encoding="utf-8") as f:
    metadata = json.load(f)

# Init GenAI client using env var GENAI_API_KEY
client = genai.Client(api_key="AIzaSyDhRYNFnCgvXSqfJZ3Xz3YX8ClNtBNvoo8")

def retrieve_contexts(query_text, top_k=3):
    q_emb = model.encode([query_text]).astype(np.float32)
    distances, indices = index.search(q_emb, top_k)
    results = []
    for rank, idx in enumerate(indices[0]):
        results.append({
            "rank": rank + 1,
            "index": int(idx),
            "answer": metadata[idx].get("answer", ""),
            "question": metadata[idx].get("question", ""),
            "distance": float(distances[0][rank])
        })
    return results

def answer_with_gemini(query_text, top_k=3, model_name="gemini-2.5-flash"):
    contexts = retrieve_contexts(query_text, top_k=top_k)
    # Build context block for the prompt
    context_block = "\n\n".join(
        f"[Source {c['rank']} | idx={c['index']} | dist={c['distance']:.4f}]\nQ: {c['question']}\nA: {c['answer']}"
        for c in contexts
    )

    prompt = (
        "You are a helpful assistant. Use ONLY the provided context to answer the question. "
        "If the context does not contain the answer, say \"I don't know\". Be concise.\n\n"
        f"CONTEXT:\n{context_block}\n\nQUESTION: {query_text}\n\nANSWER:"
    )

    resp = client.models.generate_content(
        model=model_name,
        contents=prompt
    )
    # resp.text contains generated answer (depends on SDK version)
    return {
        "generated": getattr(resp, "text", None) or resp.get("candidates",[{}])[0].get("text"),
        "contexts": contexts,
        "raw_response": resp
    }

# Example
if __name__ == "__main__":
    q = "Airport"
    out = answer_with_gemini(q, top_k=3)
    print("Generated answer:\n", out["generated"])
    print("\nRetrieved contexts:")
    for c in out["contexts"]:
        print(c)
# ...existing code...
```
